### Charged Regular Black Holes

Dado as seguintes hipóteses: (i) $8 \pi \rho_m + \frac{q^2}{r^4} = \frac{3}{R^2}(1 - \frac{r^2}{b^2})$; (ii) $\rho_m + p = 0$. Segue que $(M,g)_{int}$ fica caracterizado por:
\begin{align}
& B(r) = A^{-1}(r) = 1 - \frac{r^2}{R^2} + \frac{3 r^4}{5 R^2 b^2}, \\
& \rho_m(r) = - p(r) = \frac{3}{8 \pi R^2}(1 - \frac{3r^2}{2b^2}), \\
& q(r) = \sqrt{\frac{3}{2}} \frac{r^3}{Rb}, \ \  \rho_e(r) = \rho_{e 0} A^{-1/2}(r). 
\end{align}

Colando com $(M,g)_{ext} = (M,g)_{RN}$ no qual $B(r) = A^{-1}(r) = 1 - \frac{2m}{r} + \frac{q^2}{r^2}$ em $r = r_0$  vem que
\begin{align}
r_0 = \sqrt{\frac{2}{3}}b, \ q(r_0) = \frac{r_0^2}{R}, \ m(r_0) = \frac{4 r_0^3}{5 R^2} = \frac{4 q^2}{5 r_0}, \ \rho_m(r_0) = p(r_0) = 0.
\end{align}


Para não haver horizontes interiores, olha-se para $B^-(r_{int}) = 1/A^{-}(r_{int}) = 1 - \frac{r_{int}^2}{R^2} + \frac{3 r_{int}^4}{5 R^2 b^2} = 0$, de modo que
\begin{align}
r^2_{int} = \frac{5}{4}r_0^2(1 \pm \sqrt{1 - \frac{8}{5} \alpha^2}), \ \ \alpha = \frac{R}{r_0}.
\end{align}

Assim, vemos que $r_{int} \in \mathbb{C}$ sempre que $\alpha \ge \sqrt{\frac{5}{8}}$. Além disso, $r_{int} = r_0$ quando $\alpha = \sqrt{\frac{3}{5}}$. Logo, é possível ver que não teremos horizontes interiores sempre que $\alpha \ge \sqrt{\frac{3}{5}}$, bastando aplicar a condição de que $r_{int} \ge r_0$.


Expressando $R, m, r_0$ em termos de $q$ e $\alpha$, temos que
\begin{align}
r_{0} = \alpha q, \ R = \alpha r_0 = \alpha^2 q, \ m = \frac{4}{5 \alpha}q.
\end{align}

Os horizontes exteriores ficam caracterizados por
\begin{align}
r_{\pm} = \frac{4 q}{5\alpha}\Big(1 \pm \sqrt{1 - \frac{25}{16} \alpha^2}\Big),
\end{align}
sendo $r_+$ o horizonte de eventos e $r_-$ o horizonte de Cauchy. Para $\alpha = \sqrt{\frac{3}{5}}$ temos que $r_- = r_0$.

Além disso, podemos escrever
\begin{align}
& B^-(r) = 1 - \frac{1}{\alpha^2}(\frac{r^2}{r_0} + \frac{2 r^4}{5r_0^4}), \\
& B^+(r) = 1 - \frac{1}{\alpha^2}(\frac{8 r_0}{5 r} + \frac{r^0}{r^2}),
\end{align}
de modo que podemos plotar $B^{\pm}$ em funçao de $r_0/r$.


```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
plt.style.use(['science', 'notebook', 'grid'])
ax = fig.add_subplot()
r = np.arange(np.sqrt(3/5), 0.82, 0.00001)
x = np.arange(np.sqrt(3/5), 4/5, 0.00001)



rc = (0.8/x)*(1 - np.sqrt(1 - (25/16)*x**2))
rh = (0.8/x)*(1 + np.sqrt(1 - (25/16)*x**2))
ro = r

plt.plot(r, r, color="darkgreen", linewidth=1.5, linestyle="--", label= r'$\frac{r_0}{q}$')
plt.plot(x, rc, color="darkblue", linewidth=1.5, linestyle="-.", label= r'$\frac{r_-}{q}$')
plt.plot(x, rh, color="black", linewidth=1.5, linestyle="-.", label= r'$\frac{r_+}{q}$')

#plt.plot(x, Sl2, color="darkblue", linewidth=1.5, linestyle="-.", label= r'$\frac{r_s}{r} = 0.75$')
#plt.plot(x, Sl3, color="red", linewidth=1.5, linestyle=":", label= r'$\frac{r_s}{r} = 0.95$')

plt.xlabel(r'$\alpha$',fontsize = 18); #plt.ylabel(r'$\mathcal{V}$', fontsize=18)
plt.legend(fontsize=20)
#ax.text(0, 14, r'$m = 4, \ a = 1, \ l = 1$', fontsize=20)
#plt.savefig('horizon.pdf', format='pdf',dpi = 1200, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
plt.style.use(['science', 'notebook', 'grid'])
#ax = fig.add_subplot()
#r = np.arange(np.sqrt(5/8), 0.82, 0.00001)
x = np.arange(0, 1, 0.0001)

B0 = 1 - (1/(0.6))*((8/5)*x - x**2)

B1 = 1 - (1/(0.625))*((8/5)*x - x**2)
B2 = 1 - (1/(0.64))*((8/5)*x - x**2)
B3 = 1 - (1/(7/10))*((8/5)*x - x**2)
B4 = 1 - (1/(8/10))*((8/5)*x - x**2)

plt.plot(x, B0, color="orange", linewidth=1.5, linestyle="-", label= r'$\alpha = \sqrt{\frac{3}{5}}$')

plt.plot(x, B1, color="darkgreen", linewidth=1.5, linestyle="--", label= r'$\alpha = \sqrt{\frac{5}{8}}$')
plt.plot(x, B2, color="darkblue", linewidth=1.5, linestyle="-.", label= r'$\alpha = \sqrt{\frac{4}{5}}$')
plt.plot(x, B3, color="black", linewidth=1.5, linestyle="-.", label= r'$\alpha = \sqrt{0/7}$')
plt.plot(x, B4, color="red", linewidth=1.5, linestyle="-", label= r'$\alpha = \sqrt{0.8}$')


#plt.plot(x, Sl2, color="darkblue", linewidth=1.5, linestyle="-.", label= r'$\frac{r_s}{r} = 0.75$')
#plt.plot(x, Sl3, color="red", linewidth=1.5, linestyle=":", label= r'$\frac{r_s}{r} = 0.95$')

plt.xlabel(r'$r_0/r$',fontsize = 18); 
plt.ylabel(r'$B^+$', fontsize=18)
plt.legend(fontsize=20)
#ax.text(0, 14, r'$m = 4, \ a = 1, \ l = 1$', fontsize=20)
#plt.savefig('horizon.pdf', format='pdf',dpi = 1200, bbox_inches='tight')
plt.show()
```
