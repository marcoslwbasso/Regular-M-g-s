## Compact regular objects from a electrified Tolman-like density: An interior for Kerr-Newman spacetime

### Static 

1. Motivation for Tolman-like density and the generalization for Charged Tolman-like density. Papers: arXiv:1806.00568, arXiv:1506.05813, Possible physical realization of Tolman).
2. Tipos de objetos: Tension stars, Regular Black Holes, WH(?).
3. Limite de $b \to \infty$ = Lemos & Zanchin 
4. Condições de Energia/ Escalares de Curvatura.
5. Equivalência com as hipóteses de Tiwari.

#### Interior
Consideremos um $(M,g)$ estático e esfericamente simétrico de modo que a métrica pode ser escrita na forma
\begin{align}
ds^2 = - B(r) dt^2 + A(r) dr^2 + r^2(d \theta^2 + \sin^2 \theta d \varphi^2).
\end{align}
Consideremos também que numa certa região (interior) há um fluido perfeito e carregado que respeita as mesmas simetrias que $(M,g)$ de modo que temos 5 quantidades a serem determinadas $\{\rho_m, \rho_e, p, A, B \}$. Além disso, tal fluido satisfaz as seguintes hipóteses: 

(i) Charged Tolman like-density: $8 \pi \rho_m + \frac{q^2}{r^4} = \frac{3}{R^2}(1 - \frac{r^2}{b^2})$; 

(ii) De-Sitter: $\rho_m + p = 0$. 

A partir de $G_{\mu \nu} = 8 \pi T_{\mu \nu}$, segue que $(M,g)_{int}$ fica caracterizado por:
\begin{align}
& B(r) = A^{-1}(r) = 1 - \frac{r^2}{R^2} + \frac{r^4}{a^4}, \ \ \ \ \frac{1}{a^4} = \frac{3}{5b^2R^2}.
\end{align}

Agora, dado $\nabla_{\mu} T^{\mu \nu}=0$ e  $\nabla_{\nu} F^{\mu \nu} = 4 \pi J^{\mu}$, temos que
\begin{align}
& 2 p'(r) + \frac{2 \rho_e(r)}{\sqrt{B(r)}} \phi'(r) = 0, \\
& q(r) = 4 \pi \int \rho_e(r) \sqrt{A(r)} r^2 dr = - \frac{r^2 \phi'(r)}{\sqrt{A(r)B(r)}},
\end{align}
segue que
\begin{align}
\frac{d}{dr}p(r) =  \frac{1}{8\pi r^4} \frac{d}{dr}q^2(r).
\end{align}
Usando hipóteses (i) e (ii), chega-se em
\begin{align}
& \rho_m(r) = - p(r) = \frac{3}{8 \pi R^2}(1 - \frac{3r^2}{2b^2}), \\
& q(r) = \sqrt{\frac{3}{2}} \frac{r^3}{Rb}, \ \  \rho_e(r) = \rho_{e 0} A^{-1/2}(r), \\
& E(r) = \sqrt{\frac{3}{2}} \frac{r}{R b}.
\end{align}
Vale ressaltar que:
1. As hipóteses (i) e (ii) são equivalentes a assumir $ \rho_e(r) = \rho_{e 0} A^{-1/2}(r)$, como assumido por Tiwari (PRD, 1984). 
2. Além disso, podemos ver que no limite $b \to \infty$, obtemos a solução interior de Lemos e Zanchin (PRD, 2011), uma vez que $8 \pi \rho_m + \frac{q^2}{r^4} = \frac{3}{R^2}$. 


#### Gluing with $(M,g)_{RN}$
Como $(M,g)$ exterior, consideremos $(M,g)_{ext} = (M,g)_{RN}$ no qual $B(r) = A^{-1}(r) = 1 - \frac{2m}{r} + \frac{q^2}{r^2}$. Como estamos construindo objetos compactos, vamos colar ambos $(M,g)$ na superfície definida por $\Sigma: r = r_0$ usando as condições de junção, i.e., 
\begin{align}
& g^-_{\mu \nu}|_{r = r_0} = g^{+}_{\mu \nu}|_{r = r_0}, \ \ \ & K^-_{ab} = K^+_{ab}. 
\end{align}
As condições de junção implicam no seguinte sistema de equações:
\begin{align}
& \frac{r_0^2}{R^2}(2 - \frac{9 r_0^2}{5 b^2}) = \frac{m}{r_0}, \\  
& \frac{3 r_0^2}{R^2}(1 - \frac{r_0^2}{b^2}) = \frac{q^2}{r_0^2},
\end{align}
sendo possível observar que, no limite $b \to \infty$, recuperamos $Rq = \sqrt{3} r_0^2$ e $R^2 m = 2 r_0^2$ (Lemos e Zanchin PRD, 2011). Por fim, junto com as equações obtidas para $q(r), \rho_m(r)$ e $M(r) = 4 \pi \int T^{t}_{\ t} r^2 dr + \frac{q^2(r)}{2 r}$, vem que
\begin{align}
r_0 = \sqrt{\frac{2}{3}}b, \ q = q(r_0) = \frac{r_0^2}{R}, \ M(r_0) = \frac{4 r_0^3}{5 R^2} = \frac{4 q^2}{5 r_0}, \ \rho_m(r_0) = p(r_0) = 0.
\end{align}


Agora, analisemos os zeros de $B^-(r_{int}) = 1/A^{-}(r_{int}) = 1 - \frac{r_{int}^2}{R^2} + \frac{3 r_{int}^4}{5 R^2 b^2} = 0$, de modo que
\begin{align}
r^2_{int} = \frac{5}{4}r_0^2(1 \pm \sqrt{1 - \frac{8}{5} \alpha^2}), \ \ \alpha = \frac{R}{r_0}.
\end{align}

Assim, vemos que $r_{int} \in \mathbb{C}$ sempre que $\alpha \ge \sqrt{\frac{5}{8}}$. Além disso, $r_{int} = r_0$ quando $\alpha = \sqrt{\frac{3}{5}}$. Logo, é possível ver que não teremos horizontes interiores sempre que $\alpha \ge \sqrt{\frac{3}{5}}$, bastando aplicar a condição de que $r_{int} \ge r_0$.


Expressando $R, m, r_0$ em termos de $q$ e $\alpha$, temos que
\begin{align}
r_{0} = \alpha q, \ R = \alpha r_0 = \alpha^2 q, \ m = \frac{4}{5 \alpha}q.
\end{align}

Os horizontes exteriores ficam caracterizados por
\begin{align}
r_{\pm} = \frac{4 q}{5\alpha}\Big(1 \pm \sqrt{1 - \frac{25}{16} \alpha^2}\Big),
\end{align}
sendo $r_+$ o horizonte de eventos e $r_-$ o horizonte de Cauchy. Assim, $r_{\pm} \in \mathbb{R}$ para $\alpha \le 4/5$. Para $\alpha = \sqrt{\frac{3}{5}}$ temos que $r_- = r_0$. Assim, teremos buracos negros regulares no intervalo $\sqrt{\frac{3}{5}} \le \alpha \le \frac{4}{5}$ com $r_0 \le r_- \le r_+$. Para $\alpha > 4/5$, teremos então overcharged tension stars.  

Além disso, podemos escrever
\begin{align}
& B^-(r) = 1 - \frac{1}{\alpha^2}(\frac{r^2}{r^2_0} - \frac{2 r^4}{5r_0^4}), \\
& B^+(r) = 1 - \frac{1}{\alpha^2}(\frac{8 r_0}{5 r} - \frac{r_0}{r^2}),
\end{align}
de modo que podemos plotar $B^{\pm}$ em funçao de $r_0/r$.

#### Relacionando os potenciais métrico e elétrico

Dado
$$
B(r) = 1 - \frac{r^2}{R^2} + \frac{r^4}{a^4}, \ \ \frac{1}{a^4} = \frac{3}{5 b^2 R^2} = \frac{2}{5 r_0^2 R^2}, 
$$
podemos escrever $B(r)$ na forma
$$
B(r) = (\alpha - \beta r^2)^2 + \gamma^2, \ \ \ \alpha = \sqrt{\frac{5}{8}}\frac{r_0}{R}, \ \beta = \frac{1}{a^2} = \sqrt{\frac{2}{5}}\frac{1}{r_0 R}, \ \gamma = \sqrt{1 - \alpha^2}.
$$ 


Agora, notemos que
$$
\phi(r) = - \int \frac{q(r)}{r^2}dr = - \frac{r^2}{2 r_0 R} + \phi_0, \ \ r \le r_0
$$
Impondo que $\phi_-(r_0) = \phi_+(r_0)$, temos que
$$
\phi(r) = - \frac{q}{2 r_0}(1 + \frac{r^2}{r_0^2}).
$$
Assim, podemos escrever
$$
B(r)  = (\alpha + \sqrt{\frac{8}{5}}(\phi - \phi_0))^2 + \gamma^2  = A( \eta + \phi)^2 + \gamma^2, \ \ \ A = \frac{8}{5}, \ \eta = \sqrt{\frac{5}{8}} \alpha - \phi_0.
$$

Ou seja,
$$
B(r) = B(\phi(r)) =  A( \eta + \phi)^2 + \gamma^2.
$$

Além disso, da equação de equilíbrio, temos que
$$
2 p'(r) + \frac{2 \rho_e(r)}{\sqrt{B(r)}} \phi'(r) = 0 \implies \rho_m (r) = \rho_{e,0} \phi(r) + cte,
$$
onde $\rho_e = \rho_{e,0} A^{-1/2}(r)$.

Ou seja, também podemos ver que $B(r) = B(\rho_m(r))$ de modo que as superfícies de constante $p(r), \rho_m(r), \phi(r), B(r)$ coincidem.

Dado os seguintes parâmetros
\begin{align}
m = \frac{4q^2}{5r_0} = \frac{4r_0^3}{5R^2}, \ \ q = \frac{r_0^2}{R}, \ \ r_0 = \sqrt{\frac{2}{3}}b,
\end{align}
podemos expressar a função massa $m(r)$ como
\begin{align}
& m(r) = \frac{1}{2}\left(\frac{r^3}{R^2} - \frac{3r^5}{5 R^2b^2}\right) = \frac{r^3}{2R^2}\left(1 - \frac{2r^2} {5r^2_0} \right) = \frac{5m}{8}\left(\frac{r}{r_0} \right)^3 \left(1 - \frac{2r^2} {5r^2_0} \right), \ r \le r_0 \\
& m(r) = m - \frac{q^2}{2r}, \ r \ge r_0
\end{align}
Com isso, temos que
\begin{align}
& \Delta_-(r) = r^2 + a^2 - 2m(r)r = r^2 + a^2 - \frac{5m r}{4}\left(\frac{r}{r_0} \right)^3 \left(1 - \frac{2r^2} {5r^2_0} \right), \\
& \Delta_+(r) = r^2 + a^2 - 2mr + q^2 = r^2 + a^2 - 2mr + \frac{5 r_0 m}{4}.  
\end{align}

Normalizando, temos
\begin{align}
& \Delta_-(x) =  \frac{\Delta_-(r)}{r_0^2} = x^2 + \alpha^2 - \frac{5 \beta }{4} x^4 \left(1 - \frac{2}{5} x^2 \right), \ \ x = \frac{r}{r_0}, \alpha = \frac{a}{r_0}, \beta = \frac{m}{r_0} \\
&  \Delta_+(x) = \frac{\Delta_+(r)}{r_0^2} = x^2 + \alpha^2 - 2\beta x + \frac{5 \beta}{4}.  
\end{align}

O caso crítico que diferencia $(M,g)$ com dois horizontes de $(M,g)$ com nenhum horizonte é o caso extremo no qual $r_- = r_+ \equiv r_c$, o que implica que o discriminante $\delta$ das funções acima devem se anular. Ou seja,
\begin{align}
& \delta_-(\beta) = 125 \alpha^2 \beta^2 + (25 - 180 \alpha^2 - 108 \alpha^4) \beta - 32 = 0, \\
& \delta_+(\beta) = \beta^2 - \frac{5\beta}{4} - \alpha^2 = 0
\end{align}
que implica em
\begin{align}
& \frac{m_c}{r_0} = \beta_- = \frac{1}{250 \alpha^2} \left((180 \alpha^2 + 108 \alpha^4 - 25) + \sqrt{(180 \alpha^2 + 108 \alpha^4 - 25)^2 + 500\times32 \alpha^2} \right), \\
& \frac{m_c}{r_0} = \beta_+ = \frac{5}{8}\left(1 + \sqrt{1 + \frac{64}{25}\alpha^2} \right).
\end{align}

Além disso, para que, de fato, a solução $(M,g)_{int}$ possua horizontes, é necessário que os zeros de $\Delta_-(x)$ estejam na região $x \le 1$, enquanto que para a solução $(M,g)_{ext}$ possua horizontes, é necessário que os zeros de $\Delta_+(x)$ estejam na região $x \ge 1$. O caso em que $x = 1$ e $\Delta_-(1) = \Delta_+(1) = 0$ corresponde ao horizonte de Cauchy estando na superfície $r = r_0$. Impondo então que
\begin{align}
 & \Delta_-(1) = 1^2 + \alpha^2 - \frac{5 \beta }{4} 1^4 \left(1 - \frac{2}{5} 1^2 \right) = 0, \\
 & \Delta_-(1) =  1^2 + \alpha^2 - 2\beta 1 + \frac{5 \beta}{4} = 0,
\end{align}
nos dá 
\begin{align}
 & \beta_- = \frac{4}{3}\left(1 + \alpha^2 \right), \\
 & \beta_+ = \frac{4}{3}\left(1 + \alpha^2 \right),
\end{align}
de modo que
\begin{align}
& \Delta_-(x_h) = 0, \ \text{para} \ x_h \le 1 \ \text{se} \ \beta_- \ge \frac{4}{3}\left(1 + \alpha^2 \right), \\
& \Delta_-(x_h) = 0, \ \text{para} \ x_h \ge 1 \ \text{se} \ \beta_+ \le \frac{4}{3}\left(1 + \alpha^2 \right).
\end{align}

## Kerr-Newman outside a rotating de Sitter-type core: The rotating version of Lemos and Zanchin electrically charged solution

\begin{align}
&  \Delta_-(r) = r^2 + a^2 - \frac{m}{2r_0^3}r^4 = 0, \label{eq:horieq5}  \\
&  \Delta_+(r) = r^2 + a^2 - 2mr + \frac{3 m r_0}{2} = 0, \label{eq:horieq6}
\end{align}

The polynomial equations can be easily solved in terms of radicals and the solutions are given by
\begin{align}
 %   & x_{-\pm} = \sqrt{\frac{1}{\beta}\left(1 \pm \sqrt{1 + 2 \beta \gamma^2}\right)}, \label{eq:ihoriz} \\
 & r_{k-} = r_0 \sqrt{\frac{r_0}{m}\left(1 + (-1)^k \sqrt{1 +  \frac{2m}{r_0} \frac{a^2}{r_0^2}}\right)}, \label{eq:ihoriz} \\
%    & x_{+\pm} = \beta \left(1 \pm \sqrt{1 - \left(\frac{\gamma^2}{\beta^2} + \frac{3}{2\beta} \right)} \right). \label{eq:ehoriz}\\
& r_{k+} = m \left(1 + (-1)^k \sqrt{1 - \left(\frac{a^2}{m^2} + \frac{3r_0}{2m} \right)} \right). \label{eq:ehoriz}
\end{align}
for $k = 1,2$. It is straight forward to see that $r_{1-} \in \mathbb{C}$.
